```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("EV_data1.csv")

#  View basic info
print(df.info())
print(df.isnull().sum())
```

```python
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Make a copy
df_encoded = df.copy()

#  Clean FastCharge_KmH before encoding
df_encoded['FastCharge_KmH'] = df_encoded['FastCharge_KmH'].replace(['–', '-', '—'], '0')
df_encoded['FastCharge_KmH'] = df_encoded['FastCharge_KmH'].str.replace(' km/h', '', regex=False)
df_encoded['FastCharge_KmH'] = df_encoded['FastCharge_KmH'].astype(float)

# Columns that need encoding
cat_cols = ['Brand', 'RapidCharge', 'PowerTrain', 'PlugType', 'BodyStyle', 'Segment']


le = LabelEncoder()
for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])


print(df_encoded.head())


df_encoded.to_csv('EV_data_encoded.csv', index=False)
print("✅ Saved successfully: EV_data_encoded.csv")
```
